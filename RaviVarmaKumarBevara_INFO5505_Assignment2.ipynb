**Regression**
<font color='red'>
> It is a problem of predicting a real-valued label i.e., target in the machine learning algorithm, given an unlabeled example. Predciting the price of a product based on past year sales, such as number of purchases, reviews data, product features and so on is an example of regression.
<br />
</font>

* Logistic Regression - Rather than naming this as a regression, it should belong to a classification algorithm because it will classify the dataset into binary categories. The naming is done based on the mathematical formaulation of logistic regression. 
  * It is a case of binary classification, but can be extended to multiclass classification.
  * The same linear combination of independent variables will be applicable to this case as wll, but with as twist in dependent varibale. The y variable in here will be having two possible values.
  * The cost function assisciated with this algorithm is standard logistic function also known as "Sigmnoid Function"
  * The error rate will stay the same, minimization of emprical risk, which is mean squared error.
  * But as the result of y will be a number between 0,1 our optimization criteria will be the "maximum likelihood", which will define how likely the observatoion is according to our model.
  * But, this optimization problem has no closed form of solution, in such cases we employ gradient descent.

#### Removing Outliers

#
<font color="red">
After exploring the distributions of all the attributes which are independent, I visually suspect that there are some varibales with outliers which in future can cause the bias output. So, I decided to plot the boxplots for each variable and remove the outliers based on the z-scores.
</font>

```python
def boxplot(dataset, rows, cols):

    columns = rawdata.columns.values
    number = len(columns)

    fig, axarr = plt.subplots(rows, cols, figsize=(22, 16))

    c = 0
    for i in range(rows):
        for j in range(cols):
            sns.boxplot(x='quality', y=columns[c], data=dataset, ax=axarr[i][j])
            c += 1
            if c == (number - 1, ):
                break

boxplot(rawdata, 4, 3)
```

```python
import scipy.stats as stats

transformeddata = rawdata.copy()
z = np.abs(stats.zscore(transformeddata))
transformeddata = transformeddata[(np.abs(stats.zscore(transformeddata['residual sugar'])) < 3)]
transformeddata = transformeddata[(np.abs(stats.zscore(transformeddata['chlorides'])) < 3)]
transformeddata = transformeddata[(np.abs(stats.zscore(transformeddata['free sulfur dioxide'])) < 3)]
transformeddata = transformeddata[(np.abs(stats.zscore(transformeddata['total sulfur dioxide'])) < 3)]
transformeddata = transformeddata[(np.abs(stats.zscore(transformeddata['sulphates'])) < 3)]

transformeddata.shape
```

####Transforming all the independent variables which are skewed into new variable by applying log

#
<font color="blue">
So far, after removing outliers I can still see that some variables are skewed, so I decided to use log transformation on the variables which are skewed.
</font>

```python
transformeddata['trans residual sugar'] = np.log(transformeddata['residual sugar'])
transformeddata['trans chlorides'] = np.log(transformeddata['chlorides'])
transformeddata['trans total sulfur dioxide'] = np.log(transformeddata['total sulfur dioxide'])
```

```python
cleandata = transformeddata.drop(['residual sugar','chlorides','total sulfur dioxide'],axis=1)
```

```python
cleandata.agg(['skew', 'kurtosis']).transpose()
```

```python
cols_at_end = ['quality', 'quality level']
cleandata = cleandata[[c for c in cleandata if c not in cols_at_end] 
        + [c for c in cols_at_end if c in cleandata]]
```

```python
fig, ax = plt.subplots(figsize=(15,8))

cor = cleandata.corr()
sns.heatmap(cor, xticklabels=cor.columns, yticklabels=cor.columns, annot =True, cmap='BrBG')
cmap = sns.diverging_palette(220,20, as_cmap = True)
```

```python
cleandata.describe()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import model_selection
from sklearn import metrics
```

####Logistic Regression Model-1

```python
x = cleandata.iloc[:,:11]
y = cleandata['quality level']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#model training
logisticregressionmodel1 = LogisticRegression(solver='lbfgs',max_iter=1000)
logisticregressionmodel1.fit(x_train, y_train)
predicted_y = logisticregressionmodel1.predict(x_test)
```

```python
#testing accuaracy
logisticregressionmodel1.score(x_test,y_test)
```

```python
#training accuaracy
logisticregressionmodel1.score(x_train,y_train)
```

```python
model1accuracyscore = metrics.accuracy_score(y_test, predicted_y)
print(model1accuracyscore*100)
```

```python
#cross-validation
kfoldscores = model_selection.cross_val_score(logisticregressionmodel1, x_test, y_test, cv =10, scoring='accuracy')
kfoldscores.mean()
```

```python
#metric1 - confusion matrix
model1confusionmatrix = metrics.confusion_matrix(y_test, predicted_y, labels = logisticregressionmodel1.classes_)
confusionmatriximage = metrics.ConfusionMatrixDisplay(confusion_matrix=model1confusionmatrix,display_labels=logisticregressionmodel1.classes_)
confusionmatriximage.plot(cmap='Reds')
```

```python
#metric2 - f1 score, precision, recall
model1report = metrics.classification_report(y_test,predicted_y)
print(model1report)
```

```python
model1f1score = metrics.f1_score(y_test,predicted_y,average = 'weighted')
print(model1f1score*100)
```

```python
#metric3 - roc-auc score
fpr, tpr, thresholds = metrics.roc_curve(y_test,predicted_y)
roc_auc = metrics.auc(fpr,tpr)
roccurve = metrics.RocCurveDisplay(fpr=fpr,tpr=tpr,roc_auc=roc_auc)
roccurve.plot(color='red')

plt.show()
```
